```python
import requests

r = requests.post(
    'https://lius.cc/api/llm-rag',
    json={'q': '正一道 齋醮 演變', 'n': 10},
    timeout=10,
).json()

papers = [h for h in r['hits'] if h['type'] == 'paper']

for i, p in enumerate(papers, 1):
    print(f'@misc{{lius{i},')
    print(f'  title = {{{p["name"]}}},')
    print(f'  url   = {{{p["url"]}}},')
    print(f'  year  = {{2026}}')
    print(f'}}\n')
```
