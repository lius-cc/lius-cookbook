## 1 · Day hash

```python
import datetime

def day_hash(date_str: str) -> int:
    h = 0
    for ch in date_str:
        h = (h * 31 + ord(ch)) & 0xFFFFFFFF
    return h

today = datetime.date.today().isoformat()
print(today, '→', day_hash(today))
```

## 2 · Picker

```python
POOL = [
    {"name": "太上老君", "emoji": "☯️", "slug": "deity/太上道祖老子",
     "quotes": ["道可道，非常道；名可名，非常名。",
                "上善若水，水善利萬物而不爭。"]},
    {"name": "媽祖", "emoji": "🌊", "slug": "deity/天妃林默娘",
     "quotes": ["海不揚波，眾生平安。"]},
    # ... 15 deities total
]

def pick_for(date_str, pool):
    h = day_hash(date_str)
    d = pool[h % len(pool)]
    q = d['quotes'][(h >> 8) % len(d['quotes'])]
    return {**d, 'quote': q, 'date': date_str}

pick_for(today, POOL)
```

## 3 · Why this is share-worthy

- **Same date, same card** → friends comparing notes get matching answers
- **Zero server load** — pure client-side or static-site safe
- **Multi-quote per deity** → returning users see fresh content (different quote rotation via `h >> 8`)
