# 06 · How we compare Cross-Sect Terms

Reproduces Demo D: parallel RAG over two sects, diff side-by-side.

```python
import asyncio, aiohttp

TERMS = ['三魂七魄', '五雷', '紫府']
SECTS = ['正一', '全真']

async def fetch(session, sect, term):
    async with session.post(
        'https://lius.cc/api/llm-rag',
        json={'q': f'{sect} {term}', 'n': 3},
    ) as r:
        return sect, term, (await r.json())['hits']

async def main():
    async with aiohttp.ClientSession() as s:
        tasks = [fetch(s, sect, t) for t in TERMS for sect in SECTS]
        results = await asyncio.gather(*tasks)
        for sect, term, hits in results:
            print(f'\n[{sect}] {term}')
            for h in hits[:2]:
                print(f'  - {h["name"]}')

# await main()
```
