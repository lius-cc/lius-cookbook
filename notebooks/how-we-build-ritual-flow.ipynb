```python
import requests

RITUALS = ['三朝醮', '發奏', '安胎祈嗣']

for r_name in RITUALS:
    r = requests.post(
        'https://lius.cc/api/llm-rag',
        json={'q': f'{r_name} 流程', 'n': 8},
        timeout=10,
    ).json()
    rituals_hits = [h for h in r['hits'] if h['type'] == 'ritual']
    print(f'\n=== {r_name} ===')
    for h in rituals_hits[:5]:
        print(f'  - {h["name"]} (score {h["score"]})')

# 5/24 demo will add: timeline rendering, dharma-instrument cards,
# Zhengyi-vs-Quanzhen side-by-side diff with citation page numbers.
```
