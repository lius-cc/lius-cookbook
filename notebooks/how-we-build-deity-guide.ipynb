```python
import requests

deities = [
    ('媽祖', ['天妃', '天后', '林默娘']),
    ('城隍', ['城隍爺', '城隍境主']),
    ('關聖帝君', ['關帝', '關公']),
]

for name, aliases in deities:
    q = ' '.join([name] + aliases)
    r = requests.post(
        'https://lius.cc/api/llm-rag',
        json={'q': q, 'n': 6},
        timeout=10,
    ).json()
    print(f'\n=== {name} ===')
    for h in r['hits']:
        print(f'  [{h["type"]}] {h["name"]}')
```
