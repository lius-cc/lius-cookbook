## 1 · Data shape

Each answer is a dict with three keys:
- `steps`: 5-7 ordered worship steps
- `bring`: 4-8 items to bring (incense, flowers, fruit, etc.)
- `taboo`: 3-5 traditional taboos
- Optional: `source` — a `lius.cc/n/{slug}` reference

```python
# Example: 媽祖 × 求姻緣
mazu_love = {
    "steps": [
        "媽祖殿三炷香報名 / 生辰 / 地址",
        "稟告：希望媽祖牽引正緣（不需名字）",
        "擲筊請示是否會在 3 個月 / 半年內出現",
        "如允可祈求平安符隨身（媽祖庇佑感情平穩）",
        "離廟後留意夢境與第一面之人",
    ],
    "bring": [
        "香三炷", "粉紅或紅玫瑰一束",
        "甜湯圓供品", "求姻緣紅線（部分媽祖廟有）"
    ],
    "taboo": [
        "勿穿短褲、拖鞋、無袖、暴露衣物進廟",
        "生理期一般避免進主殿",
        "勿在廟內大聲喧嘩、拍照閃光、跨門檻",
    ],
    "source": "deity/天妃林默娘"
}

from pprint import pprint
pprint(mazu_love)
```

## 2 · Render as Markdown card

Useful for blog posts, LINE OA Flex messages, IG carousel.

```python
def render_card(deity: str, occasion: str, ans: dict) -> str:
    md = [f"## {deity} · {occasion}\n"]
    md.append("### ① 5 步驟拜法")
    md += [f"{i+1}. {s}" for i, s in enumerate(ans["steps"])]
    md.append("\n### ② 該帶什麼")
    md += [f"- {b}" for b in ans["bring"]]
    md.append("\n### ③ 禁忌")
    md += [f"- {t}" for t in ans["taboo"]]
    if ans.get("source"):
        md.append(f"\n📖 相關館藏：https://lius.cc/n/{ans['source']}")
    return "\n".join(md)

print(render_card("媽祖", "求姻緣", mazu_love))
```
