## 1 · Four dimensions

| Code | Pole A | Pole B |
| --- | --- | --- |
| D/S | 動 (Dynamic) | 靜 (Still) |
| W/T | 入世 (Worldly) | 出世 (Transcendent) |
| R/E | 理 (Rational) | 感 (Emotional) |
| L/F | 律 (Lawful) | 由 (Free) |

Each of 12 questions has 4 options; each option contributes weighted points (1 or 2) to one dimension.

```python
# Sample question structure
Q1 = {
    "q": "面對突發狀況，你會...",
    "options": [
        {"text": "立刻動手解決",   "dim": "D", "weight": 2},
        {"text": "先觀察再說",     "dim": "S", "weight": 2},
        {"text": "找人一起討論",   "dim": "S", "weight": 1},
        {"text": "本能反應出招",   "dim": "D", "weight": 1},
    ],
}
Q1
```

## 2 · Computing the 4-letter key

```python
def compute_key(scores: dict) -> str:
    return (
        ('D' if scores.get('D',0) >= scores.get('S',0) else 'S')
      + ('W' if scores.get('W',0) >= scores.get('T',0) else 'T')
      + ('R' if scores.get('R',0) >= scores.get('E',0) else 'E')
      + ('L' if scores.get('L',0) >= scores.get('F',0) else 'F')
    )

compute_key({'D': 5, 'S': 3, 'W': 4, 'T': 2, 'R': 1, 'E': 5, 'L': 3, 'F': 4})
```
